```python
from sklearn.datasets import fetch_openml
```

```python
#load dataset
mnist = fetch_openml('mnist_784', version=1)
```

```python
print(mnist)
```

```python
#shapes
X = mnist.data
y = mnist.target
print(X.shape)
print(y.shape)
```

```python
import matplotlib.pyplot as plt
X = mnist.data
y = mnist.target
first_image = X.iloc[0]
print(first_image.shape)
```

```python
#visualizing
first_image = first_image.values.reshape(28, 28)
print(first_image.shape)
plt.imshow(first_image, cmap='gray')
plt.show()
```

```python
plt.imshow(first_image, cmap='gray')
plt.title(f"Digit Label: {y[0]}")
plt.axis('off')
plt.show()
```

```python
fig, axes = plt.subplots(2, 5, figsize=(10, 5))
for i, ax in enumerate(axes.flat):

    image = X.iloc[i].values.reshape(28, 28)

    ax.imshow(image, cmap='gray')

    ax.set_title(f"Label: {y[i]}")

    ax.axis('off')
plt.tight_layout()
plt.show()
print(first_image)
```

```python
print(first_image.min())
print(first_image.max())
```

```python
#before normalization
print(X.max())
print(X.min())
```

```python
X=X/255.0
print(X.max())
print(X.min())
```

```python
#final visualization
image = X.iloc[0].values.reshape(28, 28)

print(image)
plt.imshow(image, cmap='gray')
plt.title(f"Normalized Digit: {y[0]}")
plt.axis('off')
plt.show()
```

```python
import numpy as np
unique, counts = np.unique(y, return_counts=True)
digit_counts = dict(zip(unique, counts))

print(digit_counts)
plt.figure(figsize=(10,5))

plt.bar(unique, counts)

plt.xlabel("Digits")
plt.ylabel("Count")
plt.title("MNIST Digit Distribution")

plt.show()
```

```python
# Observations

- MNIST contains 70,000 handwritten digit images.
- Each image is represented as a 28×28 grayscale image.
- Images are flattened into 784 features.
- Pixel values originally range from 0–255.
- Normalization scales values between 0–1.
- The dataset is relatively balanced across all digits.
```
